```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
def plot(img):
    plt.axis("off")
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
original=cv2.imread('/kaggle/input/natural-images/natural_images/car/car_0201.jpg')
plot(original)
edges=cv2.Canny(original,100,300) 
plot(edges)
```

```python
original = cv2.imread('/kaggle/input/four-shapes/shapes/star/591.png')
gray = cv2.cvtColor(original,cv2.COLOR_BGR2GRAY) #Input should be grayscale image
plot(original)
gray = np.float32(gray)
corners = cv2.cornerHarris(gray,2,3,0.04)
corners = cv2.dilate(corners,None)
original[corners>0.2*corners.max()]=[0,0,255]
plot(original)
```

```python
img = cv2.imread('/kaggle/input/natural-images/data/natural_images/dog/dog_0273.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plot(img)
fast =cv2.FastFeatureDetector_create(threshold=50)
kp = fast.detect(gray,None)
result = cv2.drawKeypoints(img, kp, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plot(result)
```
